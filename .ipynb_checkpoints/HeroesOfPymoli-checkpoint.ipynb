```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV
pymoli_purchases = "HeroesOfPymoli/purchase_data.csv" 
```

```python
# Read with pandas
pymoli_purchases_pd = pd.read_csv(
    pymoli_purchases, encoding="iso-8859-1", low_memory=False)
pymoli_purchases_pd.head()
```

```python
#Total Number of Players
total_players= len(pymoli_purchases_pd)

# Data frame of total_players
count_players = pd.DataFrame({"Player Count":[total_players]})
count_players
```

```python
# Purchasing Analysis: Unique Items
item_count = len((pymoli_purchases_pd["Item Name"]).value_counts())

# Purchase Analysis: Average Purchase Price
avg_price = pymoli_purchases_pd["Price"].mean()

# Purchasing Analysis: total number purchases
total_purchase = pymoli_purchases_pd["Purchase ID"].count()

# Purchasing Analysis: total revenue
total_revenue = (pymoli_purchases_pd["Price"]).sum()

# Create data frame
summary_df = pd.DataFrame({"Item Number":[item_count],
                           "Average Price":[avg_price], 
                           "Total Purchases": [total_purchase], 
                           "Total Revenue": [total_revenue]})
summary_df
```

```python
# Purchasing Analysis: 
count = pymoli_purchases_pd["Item Name"].value_counts()
item_name = (pymoli_purchases_pd["Item Name"]).unique()
sum_count_df = pd.DataFrame({"Item ID":[item_count],"Item Name":[item_name],
                           "Average Price":[avg_price], 
                           "Total Purchases": [total_purchase], 
                           "Total Revenue": [total_revenue]})
sum_count_df.head() 
```

```python
# Purchasing Analysis: total revenue
# Calculate the total reviews for the entire dataset
# Hint: use the pandas' sum() method to get the sum for each row
pymoli_purchases_pd['Total Revenue'] = pymoli_purchases_pd.iloc[:, 4:].sum(axis=1)
total_revenue = sum(pymoli_purchases_pd['Total Revenue'])
```

```python
# Group by Gender
gender_data = pymoli_purchases_pd.groupby("Gender")

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = pymoli_purchases_pd["Gender"].count()
male = pymoli_purchases_pd["Gender"].value_counts()['Male']
female = pymoli_purchases_pd["Gender"].value_counts()['Female']
non_gender_specific = total_gender - male - female

# Calculate percentage of players by gender
male_percent = (male/total_gender) * 100
female_percent = (female/total_gender) * 100
non_gender_specific_percent = (non_gender_specific/total_gender) * 100
total_gender_percent = (male_percent + female_percent + non_gender_specific_percent)



#Create Data Frame
gender_table = pd.DataFrame({"Total":total_gender, "% Total":total_gender_percent, "Male": male, "% Male": male_percent,"Female": female,"% Female": female_percent, "Non-Specific":non_gender_specific, "% Non-Specific":non_gender_specific_percent,}, index=[0])

# Format the values sorted by total count in descending order, and two decimal places for the percentage
# gender_table.format({"'% Total':'{:.2f}'"}) 
# {"% Male":"{:.2f}"}, {"% Female":"{:.2f}"},{"% Non-Specific":"{:.2f}"})
# , '% Male', '% Female', '% Non-Specific'
gender_table
```

```python
# Total purchases, Average prices & Average total by gender 
total_by_gender = gender_data["Purchase ID"].count()
avgprice_by_gender = gender_data["Price"].mean()
avgtotal_by_gender = gender_data["Price"].sum()

# Average purchase total by gender divivded by purchase count by unique shoppers
avg_purchase_per_person = avgtotal_by_gender/total_gender

# Create data frame with obtained values 
gender_purchases = pd.DataFrame({"Purchase Count": total_by_gender, 
                                    "Average Purchase Price": avgprice_by_gender,
                                    "Average Purchase Value":avgtotal_by_gender,
                                    "Avg Purchase Total per Person": avg_purchase_per_person})

# Provide index in top left as "Gender"
gender_purchases.index.name = "Gender"

# Format with currency style
gender_purchases.style.format({"Average Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

```python
# Age Demographics; Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_groups = ["10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut()
pymoli_purchases_pd["Age"] = pd.cut(pymoli_purchases_pd["Age"],age_bins, labels=age_groups)
pymoli_purchases_pd["Age"] 

# Create new data frame with the added "Age Group" and group it
age_grouped = pymoli_purchases_pd.groupby("Age")

# Calculate the numbers and percentages by age group Count total players by age category
counted_age = age_grouped["SN"].nunique()
percentage_by_age = (counted_age/total_players) * 100

# Create a summary data frame to hold the results
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": counted_age})

# Format the data frame with no index name in the corner
age_demographics.index.name = None

# Optional: round the percentage column to two decimal points
age_demographics.style.format({"Percentage of Players":"{:,.2f}"})
```

```python
# Purchases by age group
purchase_by_age = age_grouped["Purchase ID"].count()

# Average purchase price by age group 
avg_price_by_age = age_grouped["Price"].mean()

# Total purchase value by age group 
total_value_by_age = age_grouped["Price"].sum() 

# Calculate the average purchase per person in the age group 
avg_purchase_pp_by_age = (total_value_by_age/purchase_by_age)
avg_purchase_pp_by_age
# Create data frame with obtained values
age_demographics = pd.DataFrame({"Purchase Count": purchase_by_age,
                                 "Average Purchase Price": avg_price_by_age,
                                 "Total Purchase Value":total_value_by_age,
                                 "Average Purchase Total per Person": avg_purchase_pp_by_age})
age_demographics
# Format the data frame with no index name in the corner
#age_demographics.index.name = None

# Format with currency style
#age_demographics.style.format({"Average Purchase Price":"${:,.2f}",
                               #"Total Purchase Value":"${:,.2f}",
                               #"Average Purchase Total per Person":"${:,.2f}"})
```

```python
# Identify the the top spenders
top5_spenders = pymoli_purchases_pd.groupby("SN")

# Total purchases by screen name
purchase_count_spender = top5_spenders["Purchase ID"].count()

# Calculate the average purchase by name 
avg_purchase_price_spender = top5_spenders["Price"].mean()

# Calculate purchase total 
purchase_total_spender = top5_spenders["Price"].sum()

# Create data frame with obtained values
top_five = pd.DataFrame({"Purchase Count": purchase_count_spender,
                             "Average Purchase Price": avg_purchase_price_spender,
                             "Total Purchase Value":purchase_total_spender})

# Descending List
spenders = top_five.sort_values(["Total Purchase Value"], ascending=False).head()

# Format with currency style
spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

```python
# New Data Frame
items = pymoli_purchases_pd[["Item ID", "Item Name", "Price"]]

# Group the item data by item id and item name 
item_info = items.groupby(["Item ID","Item Name"])

# Count the number of times an item has been purchased 
count_of_items = item_info["Price"].count()

# Calcualte the purchase value per item 
pval_per_item = (item_info["Price"].sum()) 

# Find individual item price
item_price = pval_per_item/count_of_items

# Create data frame with obtained values
most_popular = pd.DataFrame({"Purchase Count": count_of_items, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":pval_per_item})
most_popular
# Sort in descending order to obtain top spender names and provide top 5 item names
pop_5 = most_popular.sort_values(["Purchase Count"], ascending=False).head()
pop_5
# Format with currency style
#pop_5.style.format({"Item Price":"${:,.2f}",
                                #"Total Purchase Value":"${:,.2f}"})
```

```python
# Take the most_popular items data frame and change the sorting to find highest total purchase value
pop_5 = pop_5.sort_values(["Total Purchase Value"],
                                                   ascending=False).head()
# Format with currency style
pop_5.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
